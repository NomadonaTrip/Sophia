```python
from google.adk.agents import LlmAgent
from google.adk.tools import google_search
```

```python
from dotenv import load_dotenv
#load environment variables from .env file

load_dotenv()

#access Google API key

api_key = os.getenv("GEMINI_API_KEY")
```

```python
MODEL="gemini-2.5-flash-lite"

research_agent = LlmAgent(
    model=MODEL,
    name="IdeaAgent",
    description="Asks users about their interests and identifies the areas of STEM they're interested in, and which they struggle with",
    tools=[google_search],
    disallow_transfer_to_peers=True,
)
```

```python
refiner_agent = LlmAgent(
    model=MODEL,
    name="RefinerAgent",
    description="Reviews provided areas of interest and struggle within STEM and advises them on which affordable and safe projects they might find interesting",
    instruction="Do not ask or answer any questions that will not give you insight about the user's STEM interests. Do everything to steer the conversation back to science and technology projects. Do not answer any unsafe questions. Steer the user back to STEM discussions in all your responses and answers.All the projects you suggest must be executable with everyday things. They must be safe and doable with a child's lunch money, if at all expenses will be required.",
    tools=[google_search],
    disallow_transfer_to_peers=True,
)
```

```python
root_agent = LlmAgent(
    model=MODEL,
    name="Planner",
    instruction="""You are a STEM teacher. The only thing you love more than teaching STEM classes is kids. You do everything in your power to inspre them. Your goal is to provide safe, inspirational STEP project suggestions to help children learn aspects of science they might be struggling with. You want to help them see science in the world
    1. First, use "{research_agent} to brainstorm STEM ideas in physics, chemistry or biology
    2. Then, use {refiner_agent} to identify STEM projects that strengthen the child's interests, and those that stimulate their curiosity in areas of weakness.
    3. Present the final, refined list of projects.""",
    sub_agents=[research_agent, refiner_agent],
)
```

```python
from google.adk.runners import InMemoryRunner
```

```python
runner = InMemoryRunner(agent=root_agent,
                        app_name="Planner")
print("runner created")


```
